```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, lfilter
import soundfile as sf

file_path = '/content/combined_file_thai.wav'
audio, sample_rate = sf.read(file_path)

# If the audio has multiple channels, process only the first channel
if len(audio.shape) > 1:
    audio = audio[:, 0]

def butter_bandpass(lowcut, highcut, fs, order=5):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype='band')
    return b, a

def butter_bandpass_filter(data, lowcut, highcut, fs, order=5):
    b, a = butter_bandpass(lowcut, highcut, fs, order=order)
    y = lfilter(b, a, data)
    return y

lowcut = 80.0
highcut = 3000.0

filtered_audio = butter_bandpass_filter(audio, lowcut, highcut, sample_rate)


plt.figure(figsize=(12, 6))


plt.subplot(2, 1, 1)
plt.plot(audio)
plt.title('Original Audio Signal')
plt.xlabel('Sample Number')
plt.ylabel('Amplitude')

# Plot filtered signal
plt.subplot(2, 1, 2)
plt.plot(filtered_audio)
plt.title('Filtered Audio Signal (Bandpass 80-3000 Hz)')
plt.xlabel('Sample Number')
plt.ylabel('Amplitude')

plt.tight_layout()
plt.show()

print("Playing original audio...")
display(Audio(audio, rate=sample_rate))

print("Playing filtered audio...")
display(Audio(filtered_audio, rate=sample_rate))
```

# **ANN**

```python
import torch
import torch.nn as nn
import torch.optim as optim

l1, l2, l3, = 256, 128, 64
class ANNClassifier(nn.Module):
    def __init__(self, input_dim):
        super(ANNClassifier, self).__init__()
        self.layer1 = nn.Linear(input_dim, l1)
        self.layer2 = nn.Linear(l1, l2)
        self.layer3 = nn.Linear(l2, l3)
        self.output = nn.Linear(l3, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        x = torch.relu(self.layer1(x))
        x = torch.relu(self.layer2(x))
        x = torch.relu(self.layer3(x))
        x = self.sigmoid(self.output(x))
        return x

input_dim = X_train.shape[1]

model = ANNClassifier(input_dim).to(device)

# Loss and optimizer
criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Training loop
epochs = 20
for epoch in range(epochs):
    model.train()
    running_loss = 0.0
    for inputs, labels in train_loader:
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()

    print(f'Epoch {epoch+1}/{epochs}, Loss: {running_loss/len(train_loader)}')
```

```python
from sklearn.metrics import f1_score
# Evaluate the model
model.eval()
with torch.no_grad():
    y_pred = model(X_test_tensor)
    y_pred = (y_pred > 0.5).float()
    accuracy = (y_pred == y_test_tensor).sum().item() / y_test_tensor.size(0)
    print(f'Accuracy: {accuracy}')

    # Calculate F1 score
    y_pred_np = y_pred.cpu().numpy()
    y_test_np = y_test_tensor.cpu().numpy()
    f1 = f1_score(y_test_np, y_pred_np, average='binary')
    print(f'F1 Score: {f1}')
```

```python
import torch
import numpy as np
from kan import KAN

# Define the dataset
dataset = {
    'train_input': torch.rand(100, 2),  # 100 samples, 2 features each
    'train_output': torch.rand(100, 1),  # 100 samples, 1 output each
    'test_input': torch.rand(20, 2),    # 20 samples, 2 features each
    'test_output': torch.rand(20, 1)    # 20 samples, 1 output each
}

# Create a KAN: 2D inputs, 1D output, and 5 hidden neurons, cubic spline (k=3), 5 grid intervals (grid=5)
model = KAN(width=[2, 5, 1], grid=5, k=3, seed=0)

# Train the model
model.train(dataset, opt="LBFGS", steps=20, lamb=0.01, lamb_entropy=10.)

# Save the model
torch.save(model.state_dict(), 'kan_model.pth')

# Initialize the model again
model2 = KAN(width=[2, 5, 1], grid=5, k=3, seed=0)

# Load the trained model
model2.load_state_dict(torch.load('kan_model.pth'))
model2.eval()

# Define the prediction function
def predict(model, input_data):
    model.eval()
    with torch.no_grad():
        input_tensor = torch.tensor(input_data, dtype=torch.float32)
        prediction = model(input_tensor)
    return prediction.numpy()

# Example usage
input_data = np.array([[0.5, 1.0], [1.5, 2.0]])  # Example input data
predictions = predict(model2, input_data)
print(predictions)
```

```python
import torch
import numpy as np
from kan import KAN

# Define the function
f = lambda x: torch.exp(torch.sin(torch.pi*x[:,[0]]) + x[:,[1]]**2)

# Create the dataset
def create_dataset(f, n_var, n_samples=1000):
    x = torch.rand(n_samples, n_var)
    y = f(x)
    return {'train_input': x[:800], 'train_label': y[:800], 'test_input': x[800:], 'test_label': y[800:]}

dataset = create_dataset(f, n_var=2)
print(dataset['train_input'].shape, dataset['train_label'].shape)

# Create the KAN model
model = KAN(width=[2, 5, 1], grid=5, k=3, seed=0)

# Train the model
model.train(dataset, opt="LBFGS", steps=20, lamb=0.01, lamb_entropy=10.)

# Save the model
torch.save(model.state_dict(), 'kan_model.pth')

# Initialize the model again
model2 = KAN(width=[2, 5, 1], grid=5, k=3, seed=0)

# Load the trained model
model2.load_state_dict(torch.load('kan_model.pth'))
model2.eval()

# Define the prediction function
def predict(model, input_data):
    model.eval()
    with torch.no_grad():
        input_tensor = torch.tensor(input_data, dtype=torch.float32)
        prediction = model(input_tensor)
    return prediction.numpy()

# Example usage
input_data = np.array([[0.5, 1.0], [1.5, 2.0]])  # Example input data
predictions = predict(model2, input_data)
print(predictions)
```
